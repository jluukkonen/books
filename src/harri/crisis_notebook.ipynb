```python
from sqlframe_gizmosql import GizmoSQLSession, GizmoSQLTable
import sqlframe_gizmosql.functions as F

from hereutil import here
import yaml

with open(here("db_secret.yaml"), "r") as f:
    db_params = yaml.safe_load(f)


# Create an SQLFrame (PySpark) session connected to GizmoSQL
session: GizmoSQLSession = GizmoSQLSession.builder \
    .config("gizmosql.uri", db_params["uri"]) \
    .config("gizmosql.username", db_params["username"]) \
    .config("gizmosql.password", db_params["password"]) \
    .getOrCreate()

c = F.col

# -------------------------------------------------------------------
# Tables
# -------------------------------------------------------------------

wd_entities: GizmoSQLTable = session.table("books.wd_entities")
wd_labels: GizmoSQLTable = session.table("books.wd_labels")
wd_descriptions: GizmoSQLTable = session.table("books.wd_descriptions")

wd_claim_time: GizmoSQLTable = session.table("books.wd_claim_time")
wd_claim_wikibase_entityid: GizmoSQLTable = session.table(
    "books.wd_claim_wikibase_entityid"
)

# -------------------------------------------------------------------
# Constants
# -------------------------------------------------------------------

# Great Northern War
ENTITY_ID = 21423

# Properties
P31_INSTANCE_OF = 84
P580_START_TIME = 177
P582_END_TIME = 288
P276_LOCATION = 2019

# -------------------------------------------------------------------
# Event label + description (English)
# -------------------------------------------------------------------

event_metadata = (
    wd_entities.alias("e")
    .join(
        wd_labels.alias("l"),
        (
            (c("e.entity_id") == c("l.entity_id")) &
            (c("l.language") == F.lit("en"))
        ),
        how="left"
    )
    .join(
        wd_descriptions.alias("d"),
        (
            (c("e.entity_id") == c("d.entity_id")) &
            (c("d.language") == F.lit("en"))
        ),
        how="left"
    )
    .filter(c("e.entity_id") == ENTITY_ID)
    .select(
        c("e.id").alias("qid"),
        c("l.label").alias("label"),
        c("d.description").alias("description")
    )
)

event_metadata.show()

# -------------------------------------------------------------------
# Start + end dates
# -------------------------------------------------------------------

event_times = (
    wd_claim_time
    .filter(
        (c("entity_id") == ENTITY_ID) &
        (
            c("property_id").isin(
                P580_START_TIME,
                P582_END_TIME
            )
        )
    )
    .select(
        "property_id",
        "time",
        "precision",
        "rank"
    )
)

event_times.show()

# -------------------------------------------------------------------
# Instance of
# -------------------------------------------------------------------

instance_of = (
    wd_claim_wikibase_entityid.alias("claims")

    .join(
        wd_entities.alias("target"),
        c("claims.value_entity_id") == c("target.entity_id")
    )

    .join(
        wd_labels.alias("labels"),
        (
            (c("target.entity_id") == c("labels.entity_id")) &
            (c("labels.language") == F.lit("en"))
        ),
        how="left"
    )

    .filter(
        (c("claims.entity_id") == ENTITY_ID) &
        (c("claims.property_id") == P31_INSTANCE_OF)
    )

    .select(
        c("target.id").alias("target_qid"),
        c("labels.label").alias("instance_of")
    )
)

instance_of.show()

# -------------------------------------------------------------------
# Locations
# -------------------------------------------------------------------

locations = (
    wd_claim_wikibase_entityid.alias("claims")

    .join(
        wd_entities.alias("target"),
        c("claims.value_entity_id") == c("target.entity_id")
    )

    .join(
        wd_labels.alias("labels"),
        (
            (c("target.entity_id") == c("labels.entity_id")) &
            (c("labels.language") == F.lit("en"))
        ),
        how="left"
    )

    .filter(
        (c("claims.entity_id") == ENTITY_ID) &
        (c("claims.property_id") == P276_LOCATION)
    )

    .select(
        c("target.id").alias("location_qid"),
        c("labels.label").alias("location")
    )
)

locations.show()
```

```python
from adbc_driver_flightsql import dbapi
from hereutil import here
import yaml

# -------------------------------------------------------------------
# Load DB credentials
# -------------------------------------------------------------------

with open(here("db_secret.yaml"), "r") as f:
    db_params = yaml.safe_load(f)

# -------------------------------------------------------------------
# Constants
# -------------------------------------------------------------------

# Great Northern War
ENTITY_ID = 21423

# Properties
P31_INSTANCE_OF = 84
P580_START_TIME = 177
P582_END_TIME = 288
P276_LOCATION = 2019

# -------------------------------------------------------------------
# Connect
# -------------------------------------------------------------------

with dbapi.connect(
    db_params["uri"],
    db_kwargs=dict(
        username=db_params["username"],
        password=db_params["password"]
    )
) as con:

    with con.cursor() as cur:

        # -----------------------------------------------------------
        # Event metadata (label + description)
        # -----------------------------------------------------------

        metadata_query = f"""
        SELECT
            e.id AS qid,
            l.label AS label,
            d.description AS description
        FROM books.wd_entities e

        LEFT JOIN books.wd_labels l
            ON e.entity_id = l.entity_id
           AND l.language = 'en'

        LEFT JOIN books.wd_descriptions d
            ON e.entity_id = d.entity_id
           AND d.language = 'en'

        WHERE e.entity_id = {ENTITY_ID}
        """

        print("\n=== EVENT METADATA ===")
        cur.execute(metadata_query)
        print(cur.fetch_arrow_table())

        # -----------------------------------------------------------
        # Start + end dates
        # -----------------------------------------------------------

        dates_query = f"""
        SELECT
            property_id,
            time,
            precision,
            rank
        FROM books.wd_claim_time
        WHERE entity_id = {ENTITY_ID}
          AND property_id IN ({P580_START_TIME}, {P582_END_TIME})
        """

        print("\n=== EVENT DATES ===")
        cur.execute(dates_query)
        print(cur.fetch_arrow_table())

        # -----------------------------------------------------------
        # Instance of
        # -----------------------------------------------------------

        instance_query = f"""
        SELECT
            target.id AS target_qid,
            labels.label AS instance_of

        FROM books.wd_claim_wikibase_entityid claims

        JOIN books.wd_entities target
            ON claims.value_entity_id = target.entity_id

        LEFT JOIN books.wd_labels labels
            ON target.entity_id = labels.entity_id
           AND labels.language = 'en'

        WHERE claims.entity_id = {ENTITY_ID}
          AND claims.property_id = {P31_INSTANCE_OF}
        """

        print("\n=== INSTANCE OF ===")
        cur.execute(instance_query)
        print(cur.fetch_arrow_table())

        # -----------------------------------------------------------
        # Locations
        # -----------------------------------------------------------

        location_query = f"""
        SELECT
            target.id AS location_qid,
            labels.label AS location

        FROM books.wd_claim_wikibase_entityid claims

        JOIN books.wd_entities target
            ON claims.value_entity_id = target.entity_id

        LEFT JOIN books.wd_labels labels
            ON target.entity_id = labels.entity_id
           AND labels.language = 'en'

        WHERE claims.entity_id = {ENTITY_ID}
          AND claims.property_id = {P276_LOCATION}
        """

        print("\n=== LOCATIONS ===")
        cur.execute(location_query)
        print(cur.fetch_arrow_table())
```

```python
from adbc_driver_flightsql import dbapi
from hereutil import here
import yaml

# -------------------------------------------------------------------
# Load DB credentials
# -------------------------------------------------------------------

with open(here("db_secret.yaml"), "r") as f:
    db_params = yaml.safe_load(f)

# -------------------------------------------------------------------
# Input war name
# -------------------------------------------------------------------

war_name = "Great Northern War"

# -------------------------------------------------------------------
# Connect
# -------------------------------------------------------------------

with dbapi.connect(
    db_params["uri"],
    db_kwargs=dict(
        username=db_params["username"],
        password=db_params["password"]
    )
) as con:

    with con.cursor() as cur:

        # -----------------------------------------------------------
        # Find matching Wikidata entities by English label
        # -----------------------------------------------------------

        query = f"""
        SELECT
            e.entity_id,
            e.id AS wikidata_id,
            l.label,
            d.description

        FROM books.wd_labels l

        JOIN books.wd_entities e
            ON l.entity_id = e.entity_id

        LEFT JOIN books.wd_descriptions d
            ON e.entity_id = d.entity_id
           AND d.language = 'en'

        WHERE l.language = 'en'
          AND lower(l.label) = lower('{war_name}')

        LIMIT 20
        """

        cur.execute(query)

        result = cur.fetch_arrow_table()

        print(result)
```
